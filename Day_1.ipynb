```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

a=np.arange(40,50)
b=np.arange(50,60)
plt.plot(a,b,'b*--',linewidth=1,markersize=10)
plt.title("Line Plot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
